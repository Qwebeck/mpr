```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
one_node_send = np.fromfile('output_one_node_MPI_Send.txt', dtype=float, count=-1, sep='\n', offset=0)
two_nodes_send = np.fromfile('output_two_nodes_MPI_Send.txt', dtype=float, count=-1, sep='\n', offset=0)
one_node_bsend = np.fromfile('output_one_node_MPI_BSend.txt', dtype=float, count=-1, sep='\n', offset=0)
two_nodes_bsend = np.fromfile('output_two_nodes_MPI_BSend.txt', dtype=float, count=-1, sep='\n', offset=0)

bit_size = np.arange(len(one_node_send)) * 8
```

```python
sns.set_theme(style="darkgrid")
font = {'size' : 10 }
matplotlib.rc('font', **font)
plt.xlabel('Data size')
plt.ylabel('MBi / s')
plt.scatter(x=bit_size, y=bit_size / one_node_send, s=1, label =   'Send: one node.')
plt.scatter(x=bit_size, y=bit_size / two_nodes_send, s=1, label =  'Send: two nodes.')
plt.scatter(x=bit_size, y=bit_size / one_node_bsend, s=1, label =  'Bsend: one node.')
plt.scatter(x=bit_size, y=bit_size / two_nodes_bsend, s=1, label = 'Bsend: two nodes.')
plt.legend(fontsize=10, title_fontsize=15)
plt.show()
```

```python
print(f'Latency (one_node_send): {(one_node_send)[0]:.5f}')
print(f'Latency (two_nodes_send): {(two_nodes_send)[0]:.5f}')
print(f'Latency (one_node_bsend): {(one_node_bsend)[0]:.5f}')
print(f'Latency (two_nodes_bsend): {(two_nodes_bsend)[0]:.5f}')
```
